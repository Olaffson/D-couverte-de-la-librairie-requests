```python
# Clé API OpenWeatherMap
api_key = '3705ffb386db088afe32be3c44bb259c'

# Liste des villes à analyser
villes = ["Paris", "Marseille", "Lyon", "Toulouse", "Nice", "Nantes", "Strasbourg", "Montpellier", "Bordeaux", "Lille", "Rennes", "Reims", "Le Havre", "Saint-Étienne", "Toulon", "Grenoble", "Dijon", "Angers", "Nîmes", "Villeurbanne"]

# Fonction pour convertir le timestamp en heure/minute/seconde
def timestamp_to_time(timestamp):
    return datetime.fromtimestamp(timestamp).strftime('%H:%M:%S')

# Fonction pour convertir les degres de la direction du vent en boussole
def degrees_to_cardinal(degrees):
    """
    Convertit les degrés de la direction du vent en coordonnées d'une boussole.
    """
    cardinals = ["N", "NNE", "NE", "ENE", "E", "ESE", "SE", "SSE", "S", "SSO", "SO", "OSO", "O", "ONO", "NO", "NNO"]
    index = int((degrees / 22.5) + 0.5)
    return cardinals[index % 16]

# Initialisation du DataFrame
df = pd.DataFrame(columns=['Ville', 'Temperature', 'Temperature ressentie', 'Temperature minimale', 'Temperature maximale', 'Pression atmospherique', 'Humidite', 'Vitesse du vent', 'Direction du vent', 'Lever du soleil', 'Coucher du soleil'])

# Boucle pour récupérer les informations pour chaque ville
for ville in villes:
    url = f"http://api.openweathermap.org/data/2.5/weather?q={ville},fr&units=metric&appid={api_key}"
    response = requests.get(url)
    data = response.json()
    # Récupération des informations
    temperature = data['main']['temp']
    temperature_res = data['main']['feels_like']
    temp_min = data['main']['temp_min']
    temp_max = data['main']['temp_max']
    pression = data['main']['pressure']
    humidite = data['main']['humidity']
    vent_vitesse = data['wind']['speed']
    # direction du vent en degres
    # vent_direction = data['wind']['deg']
    # direction du vent en boussole
    vent_direction = degrees_to_cardinal(data['wind']['deg'])
    lever_soleil = timestamp_to_time(data['sys']['sunrise'])
    coucher_soleil = timestamp_to_time(data['sys']['sunset'])
    # Ajout des informations dans le DataFrame
    df = df.append({'Ville': ville, 'Temperature': temperature, 'Temperature ressentie': temperature_res, 'Temperature minimale': temp_min, 'Temperature maximale': temp_max, 'Pression atmospherique': pression, 'Humidite': humidite, 'Vitesse du vent': vent_vitesse, 'Direction du vent': vent_direction, 'Lever du soleil': lever_soleil, 'Coucher du soleil': coucher_soleil}, ignore_index=True)

# Exportation du DataFrame dans un fichier CSV
df.to_csv("informations_meteo.csv", index=False)
```
